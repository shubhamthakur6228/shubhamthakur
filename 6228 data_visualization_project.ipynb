```python

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Extract the Rank by most visits
visits_viz_rank=df['Visits']

# Descriptive statistics
mean = visits_viz_rank.mean()
median = visits_viz_rank.median()
mode = visits_viz_rank.mode()[0]
std_dev = visits_viz_rank.std()

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Standard Deviation: {std_dev}")

import matplotlib.pyplot as plt
import seaborn as sns

# Generate a count plot for visits count plt.figure(figsize=(10, 6))
sns.countplot(x=df['Visits'], color='red')
plt.title("visits_viz_rank count Plot")
plt.xlabel("Rank by most visits")
plt.ylabel("Count")
plt.grid(True)
plt.show()

# Generate a histogram
plt.figure(figsize=(20,8))
plt.hist(visits_viz_rank, bins=6, color='red', edgecolor='green')
plt.title("visits_viz_rank count Plot")
plt.xlabel("visits")
plt.ylabel("Frequency")
plt.show()
plt.hist(visits_viz_rank, bins=6, color='red', edgecolor='green')
plt.title("visits_viz_rank count Plot")
plt.xlabel("visits")
plt.ylabel("Frequency")
plt.show()

```

```python
# Scatter Plot between likes and comments
plt.figure(figsize=(15, 10))
sns.scatterplot(x='Likes', y='Comments',hue="Country", data=df, color='skyblue')
plt.title("Scatter Plot: Data Visualization vs. Python")
plt.xlabel("Data Visualization Scores")
plt.ylabel("Python Scores")
plt.legend(loc='center')
plt.show()
```

```python
# Scatter Plot between country and most visits
plt.figure(figsize=(30, 15))
sns.swarmplot(x='Country',y="Visits", hue='Categories', data=df, color='skyblue')
plt.title("Scatter Plot:Country vs. Visits")
plt.xlabel("counties")
plt.ylabel("Visits count")
plt.show()
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Generate a bar plot with hue (Country and Visits)
plt.figure(figsize=(30,15))
sns.barplot(x='Country',y="Visits", hue='Categories', data=df, ci=None, palette='Set2')
plt.title("Distribution of visits with country")
plt.xlabel("countries")
plt.ylabel("visits Count")
plt.legend(loc='upper right')
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the correlation matrix
corr_matrix = df[['Suscribers','Visits','Likes','Comments']].corr()
# Generate a heatmap
plt.figure(figsize=(15,9))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# Calculate variance of each column
variance = df.var()
print("Variance:")
print(variance)

# Calculate the covariance matrix
covariance_matrix = df.cov()
print("\nCovariance Matrix:")
print(covariance_matrix)
```

```python
# Pair Plot for Categories  Suscribers  Country Visits  Likes Comments
sns.pairplot(df[['Categories','Suscribers','Country','Visits','Likes','Comments']], diag_kind='kde')
plt.suptitle("Pair Plot of Subjects", y=1.02)
plt.show()
```
